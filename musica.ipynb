```python
import pandas as pd

#  carga de Dataset
df = pd.read_csv('/content/spotify_data clean.csv')

print("HEAD:")
print(df.head(), "\n")

print("TAIL:")
print(df.tail(), "\n")

print("SHAPE:")
print(df.shape, "\n")

print("INFO:")
print(df.info(), "\n")

print("DESCRIBE:")
print(df.describe(), "\n")
```

```python
#  Estadísticas

pop_mean = df['track_popularity'].mean()
artist_pop_max = df['artist_popularity'].max()
duration_mean = df['track_duration_min'].mean()

print("Popularidad promedio de canciones:", pop_mean)
print("Popularidad máxima del artista:", artist_pop_max)
print("Duración promedio (minutos):", duration_mean, "\n")
```

```python
# Crear columnas nuevas

df['duration_category'] = pd.cut(
    df['track_duration_min'],
    bins=[0, 2, 3.5, 10],
    labels=['Corta', 'Media', 'Larga']
)
```

```python
# Clasificación por popularidad
df['popularity_level'] = df['track_popularity'].apply(
    lambda x: 'Alta' if x >= 50 else 'Media' if x >= 20 else 'Baja'
)

print("COLUMNAS NUEVAS:")
print(df[['track_name','track_duration_min','duration_category','track_popularity','popularity_level']].head(), "\n")
```
