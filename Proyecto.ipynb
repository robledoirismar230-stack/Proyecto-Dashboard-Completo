```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

# Datos
productos = ["Manzanas", "Uvas", "Peras", "Bananos"]
ventas = [70, 58, 63, 90]

fig, ax = plt.subplots(2, 3, figsize=(15, 8))

ax_flat = ax.flatten()

configuraciones = [
    {'tipo': 'bar', 'color': 'skyblue', 'titulo': 'Vertical (Cielo)'},
    {'tipo': 'barh', 'color': 'lightcoral', 'titulo': 'Horizontal (Coral)'},
    {'tipo': 'bar', 'color': 'lightgreen', 'titulo': 'Vertical (Verde)'},
    {'tipo': 'barh', 'color': 'gold', 'titulo': 'Horizontal (Oro)'},
    {'tipo': 'bar', 'color': 'grey', 'titulo': 'Vertical (Gris)'},
    {'tipo': 'barh', 'color': 'plum', 'titulo': 'Horizontal (Ciruela)'},
]

for i, axi in enumerate(ax_flat):
    config = configuraciones[i]

    if config['tipo'] == 'bar':
        axi.bar(productos, ventas, color=config['color'])
        axi.set_ylabel("Ventas")
        axi.set_xlabel("Productos")
        axi.grid(axis='y', alpha=0.5)
    else: # Grfico de barras horizontal
        axi.barh(productos, ventas, color=config['color'])
        axi.set_xlabel("Ventas")
        axi.set_ylabel("Productos")
        axi.grid(axis='x', alpha=0.5)

    axi.set_title(config['titulo'], fontsize=12)

plt.suptitle("Distribucin de Ventas por Producto en Subplots 2x3", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
alturas = [1.50, 1.55, 1.60, 1.65, 1.70, 1.75, 1.80]
pesos = [50, 55, 65, 60, 70, 75, 80]
colores = ['blue', 'red', 'green', 'purple', 'orange', 'black']
tamanos = [50, 100, 150, 200, 250, 300]

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))

ax_flat = ax.flatten()

for i, axi in enumerate(ax_flat):

    axi.scatter(alturas, pesos,
                color=colores[i],
                s=tamanos[i],
                alpha=0.7)

    axi.set_title(f"Grfico {i+1} ({colores[i].capitalize()})", fontsize=12)
    axi.set_xlabel("Altura (h)")
    axi.set_ylabel("Peso (kg)")
    axi.grid(True, alpha=0.3)

plt.suptitle("Visualizacin de Altura vs. Peso en Subplots 2x3", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
categorias = ["Panama", "Veraguas", "Cocle", "Bocas del toro"]
valores = [30, 15, 3, 1]
# Para que una rebanada (Panam) se separe del resto en algunos subplots
explotar = (0.1, 0, 0, 0)
# Definicin de colores
colores = ['#4CAF50', '#FF9800', '#2196F3', '#9C27B0'] # Colores con un estilo de dashboard

# 1. Crear la figura y los 6 ejes (3 filas, 2 columnas)
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))

# 2. Aplanar el array de ejes para iterar fcilmente
ax_flat = ax.flatten()

# 3. Definir las configuraciones para cada uno de los 6 subplots (el "dashboard")
configuraciones = [
    # Fila 1
    {'titulo': 'Grfico 1: Porcentaje con Sombra', 'explode': None, 'shadow': True},
    {'titulo': 'Grfico 2: Separacin (Panam)', 'explode': explotar, 'shadow': False},
    # Fila 2
    {'titulo': 'Grfico 3: Inicio en 90', 'startangle': 90, 'explode': None, 'shadow': False},
    {'titulo': 'Grfico 4: Sin Etiquetas', 'labels': None, 'explode': None, 'shadow': False},
    # Fila 3
    {'titulo': 'Grfico 5: Texto Blanco', 'autopct_color': 'white', 'explode': None, 'shadow': False},
    {'titulo': 'Grfico 6: Borde Rojo', 'wedgeprops': {'linewidth': 1, 'edgecolor': 'red'}, 'explode': None, 'shadow': False}
]

# 4. Iterar sobre cada eje (subplot) para dibujar el grfico
for i, axi in enumerate(ax_flat):
    config = configuraciones[i]

    # Parmetros bsicos
    params = {
        'x': valores,
        'labels': categorias,
        'autopct': '%1.1f%%',
        'colors': colores,
        'explode': config.get('explode'),
        'shadow': config.get('shadow', False),
        'startangle': config.get('startangle', 0),
        'wedgeprops': config.get('wedgeprops', {}),
        'textprops': {'color': config.get('autopct_color', 'black')}
    }

    # Dibujar el grfico de pastel en el eje actual (axi)
    axi.pie(**params)

    # Configurar ttulo del subplot
    axi.set_title(config['titulo'], fontsize=11)

# 5. Ttulo principal y ajuste de diseo
plt.suptitle("Dashboard de Provincias Visitadas", fontsize=18, y=1.02)
plt.tight_layout()
plt.show()
```
